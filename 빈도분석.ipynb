위반내용과 단속동 빈도분석 => 각 단속동에서 어떤 위반내용이 많이 나왔는지 파악

```python
### 단속동 전처리 전
import pandas as pd

# '단속동'과 '위반내용' 변수 선택
df_subset = df[['단속동', '위반내용']]

# 결측치 제거
df_subset = df_subset.dropna()

# 각 단속동에서 어떤 위반내용이 많이 나왔는지 빈도분석 수행
violation_counts = df_subset.groupby(['단속동', '위반내용']).size().reset_index(name='counts')

# 결과 출력
print(violation_counts)
```

```python
### 단속동 전처리 ###

# 단속동 컬럼에서 공백 제거
df['단속동'] = df['단속동'].str.replace(' ', '')

unique_dong = df['단속동'].unique()
print(unique_dong)
print(len(unique_dong))

# 단속장소와 위반내용 컬럼에서 공백 제거
df['단속동'] = df['단속동'].str.replace(' ', '')

import re
# 정규 표현식 : 동일로26길 -> 동일로
pattern = r'(.로).+길'
# 주소 변경
for i in range (len(df['단속동'])):
  df['단속동'][i] = re.sub(pattern, r'\1',df['단속동'][i])

print(df)

# 단속동 count == 1 -> 35개
location_count = df['단속동'].value_counts()
countis1 = location_count[location_count==1]
#print(countis1)
countis1.shape

# location_count가 1인 행 제거
for index, value in countis1.items():
    df = df.drop(df[df['단속동'] == index].index)
```
