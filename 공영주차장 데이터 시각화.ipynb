### 지도 시각화(단속동 빈도 분석)

```python
import folium
from folium.plugins import MarkerCluster
```

```python
# 데이터 불러오기
where = "/content/drive/MyDrive/[03팀] 김가경, 김수민, 민채원/code/GwangjinGu_latilong.csv"
# 데이터 불러오기
data = pd.read_csv(where, encoding='utf-8')
data = data[data['시군구'] == '광진구']
data = data.rename(columns={'읍면동/구': '단속동'})
data
```

```python
# 단속동df, 위도/경도df -> merge
merged_df = pd.merge(data,location_count, on='단속동', how='left')
merged_df = merged_df[['단속동','단속건수','위도','경도']]
merged_df = merged_df.dropna(subset=['단속건수'])
merged_df = merged_df.sort_values(by = '단속건수', ascending = False)
merged_df['단속건수'] = merged_df['단속건수'].astype(int)
merged_df = merged_df.reset_index(drop=True) #인덱스 0부터 시작하게
merged_df
```

```python
from collections import defaultdict
import time
# 광진구 좌표
GwangjinGu = merged_df.loc[:,['위도','경도']]
lst = GwangjinGu.values.tolist() # DF -> list

# 광진구 중심 좌표
center = [ 37.538617,127.082375]
m = folium.Map(location=center,titles = 'GwangjinGu', zoom_start=13)

# 단속건수 범위 구하기
min_count = merged_df['단속건수'].min()
max_count = merged_df['단속건수'].max()
diff_count = max_count - min_count

# 원을 표시하고 빈도수에 따라 원의 크기를 조절
for i in range(len(merged_df)):
    count = merged_df.loc[i, '단속건수']
    normalized_count = (count - min_count) / diff_count  # 단속건수를 0~1 사이로 정규화
    radius = normalized_count * 80 + 5  # 원의 크기는 빈도수와 비례하도록 조절
    color = f'#{int(255*normalized_count):02x}0000'  # 원의 색상은 빈도수에 따라 변화하도록 조절
    folium.CircleMarker(location=lst[i],
                        radius=radius,
                        color=color,
                        fill_color=color).add_to(m)
    folium.Marker(location=lst[i],  # 동 이름 표시
                  icon=folium.Icon(color='green',icon='mapbox'),
                  tooltip=merged_df.loc[i, '단속동']).add_to(m)

m.save('map.html')
m
```

### 인구수 시각화

```python
import pandas as pd
# 데이터 불러오기
filename = "/content/drive/MyDrive/[03팀] 김가경, 김수민, 민채원/code/Gwangjin-Gu Population of March.csv"
# 데이터 불러오기
population = pd.read_csv(filename, encoding='cp949')
print(population)
```

```python
population['합계'] = population['합계'].apply(lambda x: int(x.replace(',', '')))
population['남'] = population['남'].apply(lambda x: int(x.replace(',', '')))
population['여'] = population['여'].apply(lambda x: int(x.replace(',', '')))
```

```python
# 행정기관 전처리 : 앞 두글자가 같은 경우 그룹화
population['행정기관'] = population['행정기관'].str[:2]
grouped_population = population.groupby('행정기관', as_index=False).sum()
grouped_population = grouped_population.sort_values(by='합계',ascending = False)
grouped_population = grouped_population.drop([1])
# 결과 확인
print(grouped_population)
```

```python
import matplotlib.pyplot as plt
colors = ['#BBDEFB','#FFD1DC']
grouped_population = grouped_population.set_index('행정기관')
ax = grouped_population[['남', '여']].plot(title = '행정기관 별 인구수',kind='bar', color = colors, stacked=True)
ax.set_xlabel('행정기관')
ax.set_ylabel('인구수')
plt.show()
```
