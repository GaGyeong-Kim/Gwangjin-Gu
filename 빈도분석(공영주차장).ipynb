```python
import csv
import codecs

filename = "/content/drive/MyDrive/[03팀] 김가경, 김수민, 민채원/code/Gwangjin-gu Public Parking lot.csv"

with codecs.open(filename, 'r', 'euc_kr', errors='ignore') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

동명 범주화 => 구의1동, 2동, 3동은 구의동으로 전처리

```python
df1.loc[df1['동명'] == '구의1동', '동명'] = 1
df1.loc[df1['동명'] == '구의2동', '동명'] = 1
df1.loc[df1['동명'] == '구의3동', '동명'] = 1
df1.loc[df1['동명'] == '중곡1동', '동명'] = 2
df1.loc[df1['동명'] == '중곡2동', '동명'] = 2
df1.loc[df1['동명'] == '중곡3동', '동명'] = 2
df1.loc[df1['동명'] == '중곡4동', '동명'] = 2
df1.loc[df1['동명'] == '능동', '동명'] = 3
df1.loc[df1['동명'] == '자양1동', '동명'] = 4
df1.loc[df1['동명'] == '자양2동', '동명'] = 4
df1.loc[df1['동명'] == '자양3동', '동명'] = 4
df1.loc[df1['동명'] == '자양4동', '동명'] = 4
df1.loc[df1['동명'] == '광장동', '동명'] = 5
df1.loc[df1['동명'] == '화양동', '동명'] = 6
df1.loc[df1['동명'] == '군자동', '동명'] = 7

df1
```

```python
print(df1['동명'].unique())
```

```python
for element in df1['동명'].unique():
    print(element)
```

광진구 동별로 공영주차장 소계 파악 => 빈도분석 진행

```python
# 동별 주차장 소계 파악하기 => 변수 두개 선택
df_subset = df1[['동명', '주차면수 소계']]

# 결측치 제거
df_subset = df_subset.dropna()

# 각 단속동에서 공영주차장 개수가 몇개인지 빈도분석 수행
parkinglot_counts = df_subset.groupby(['동명', '주차면수 소계']).size().reset_index(name='counts')

# 결과 출력
print(parkinglot_counts)
```

동별로 주차면수 소계 합한 데이터

```python
df_subset = df1[['동명', '주차면수 소계']]
df_subset = df_subset.dropna()
parkinglot_counts = df_subset.groupby('동명')['주차면수 소계'].sum().reset_index(name='합계')
print(parkinglot_counts)
```

```python
parkinglot_counts.to_csv('parkingplot.csv',index=False)
```

시각화 => x축 label 달아둠

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터프레임에서 각 동별 주차면수 소계의 합을 구합니다.
df_subset = df1[['동명', '주차면수 소계']]
df_subset = df_subset.dropna()
parkinglot_counts = df_subset.groupby('동명')['주차면수 소계'].sum().reset_index(name='합계')

# 최대값을 가지는 막대의 인덱스를 찾습니다.
max_index = parkinglot_counts['합계'].idxmax()

# 막대 그래프를 그립니다.
barplot = sns.barplot(x='동명', y='합계', data=parkinglot_counts, color='#87CEEB')

# 최대값을 가지는 막대만 빨간색으로 표시합니다.
barplot.get_children()[max_index].set_color('#CD5C5C')

plt.title('동별 공영주차장 수')
plt.xlabel('동명')
plt.ylabel('주차면수 소계 합계')

# x축 레이블을 추가합니다.
labels = ['구의동', '중곡동', '능동', '자양동', '광장동', '화양동', '군자동']
plt.xticks(range(len(labels)), labels)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

# 동별 자동차 등록대수 합계 구하기
df_subset = df2[['읍면동명', '자동차등록대수']]
df_subset = df_subset.dropna()
parkinglot_counts = df_subset.groupby('읍면동명')['자동차등록대수'].sum().reset_index(name='합계')

# 픽토그램 이미지 로딩
img = mpimg.imread('/content/drive/MyDrive/[03팀] 김가경, 김수민, 민채원/code/pictogram.png')

# 이미지 크기 및 색상 설정
fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('white')
plt.imshow(img, extent=[0, len(parkinglot_counts), 0, 10], aspect='auto', alpha=0.5)

# 데이터 포인트 그리기
colors = ['#FFD700', '#CD5C5C', '#D3D3D3', '#87CEFA', '#ADD8E6', '#98FB98', '#FFA07A']
for i in range(len(parkinglot_counts)):
    x = i + 0.5
    y = 5
    s = np.sqrt(parkinglot_counts['합계'][i]) * 5
    plt.scatter(x, y, s=s, c=colors[i], alpha=1)
    plt.annotate(parkinglot_counts['읍면동명'][i], (x, y + 0.25), ha='center', va='bottom')

# 축 레이블 설정
plt.xlim(0, len(parkinglot_counts))
plt.ylim(0, 10)
plt.xticks(np.arange(0.5, len(parkinglot_counts)), parkinglot_counts['읍면동명'], fontsize=10)
plt.yticks([], [])
plt.xlabel('동', fontsize=12)
plt.title('동별 자동차등록대수 픽토그램', fontsize=16)

plt.show()
```
